```python
from random_word import RandomWords
import random
from wordfreq import zipf_frequency

def char_in_word(c, word):
    for i in word:
        if i == c:
            return True
    
    return False

def is_all_lower(word):
    for i in word:
        if i < 'a' or i > 'z':
            return False
    return True
            

def num_chars(c, word):
    count = 0
    for i in word:
        if i == c:
            count += 1

    return count

def hit_blow_neither(guess, word):
    d = {}
    for i in set(word):
        d[i] = num_chars(i,word)
    l = ["N" for i in range(len(word))]
    for i in range(len(guess)):
        if guess[i] not in d:
            l[i] = "N"
        elif d[guess[i]] == 0:
            l[i] = "N"
        elif guess[i] == word[i]:
            l[i] = "H"
            d[guess[i]] -= 1
        elif guess[i] != word[i] and char_in_word(guess[i],word) == True:
            l[i] = "B"
            d[guess[i]] -= 1
    return l
    
def end_game_conditions(word,guess,l,round_number, rounds,give_up):
    if guess == give_up:
        return True
    if round_number == rounds:
        if word == guess:
            print("Congrats you got it!")
        return True
    for i in l:
        if i != "H":
            return False
    
    print("Congrats you got it!")
    return True
    
def not_all_letters(word):
    if type(word) == "NoneType" or word == None:
        return True
    word = word.lower()
    for i in word:
        if i < 'a' or i > 'z':
            return True
    
    return False
       
def has_duplicates(word):
    if type(word) == "NoneType" or word == None:
        return False
    d = {}
    word = word.lower()
    for i in word:
        if i in d:
            return True
        else:
            d[i] = 1
    
    return False
    

def main():
    round_count = {5: 6, 6: 8}
    r = RandomWords()
    rounds = 10
    round_num = 1
    disable_duplicates = True
    reshuffle = False
    use_custom = False
    
    pos = "noun"
    
    give_up = "GIVE UP"
    
    duplicate_status = "Duplicates Disabled" if disable_duplicates == True else "Duplicates Enabled"
    instructions = "Print GIVE UP to end the game"
    
    MIN_LEN = 5
    MAX_LEN = 6
    
    zipf_freq = 4.35
    
    word = "kjlsdaf"
    while ((word == "kjlsdaf") or (type(word) == "NoneType") or (disable_duplicates and (has_duplicates(word) == True)) or reshuffle or not_all_letters(word)) and (use_custom == False):
        word = r.get_random_word(hasDictionaryDef="true",includePartOfSpeech="noun",minLength=MIN_LEN, maxLength=MAX_LEN)
        if not ((word == None) or (type(word) == "NoneType")):
            z = zipf_frequency(word,'en')
            if z < zipf_freq:
                reshuffle = True
            else:
                reshuffle = False

    if word not in custom_words and word != "kjlsdaf":
        custom_words[word.lower()] = 1
    guess = ""
    hits_and_blows = []
    if use_custom == True:
        custom_list = list(custom_words.keys())
        word = custom_list[random.randint(0,len(custom_list))]
    
    
    if len(word) in round_count:
        rounds = round_count[len(word)]
    else:
        rounds = len(word)*2
    
    print(str(len(word))+" letters, "+str(rounds)+" rounds, "+duplicate_status)
    print(instructions)
    word = word.lower()
    #print(word)
    while round_num <= rounds:
        while (len(guess) != len(word)) or (is_all_lower(guess) == False):
            guess = str(input("Round "+str(round_num)+" of "+str(rounds)+": Please input guess: "))
            if guess == give_up:
                break
            if ((len(guess) != len(word)) or (is_all_lower(guess) == False)):
                print("Please enter a valid guess (must have valid characters and be all lowercase): ")
        
        if guess != give_up:
            hits_and_blows = hit_blow_neither(guess,word)

            print([char for char in guess])
            print(hits_and_blows)
        
        if end_game_conditions(word,guess,hits_and_blows, round_num, rounds,give_up) == True:
            print("The Game has ended")
            print("The word was: "+word)
            break
        round_num += 1
        guess = ""
```
