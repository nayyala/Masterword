```python
from random_word import RandomWords
import random
from wordfreq import zipf_frequency

def char_in_word(c, word):
    for i in word:
        if i == c:
            return True
    
    return False

def is_all_lower(word):
    for i in word:
        if i < 'a' or i > 'z':
            return False
    return True
            

def num_chars(c, word):
    count = 0
    for i in word:
        if i == c:
            count += 1

    return count

def hit_blow_neither(guess, word):
    d = {}
    for i in set(word):
        d[i] = num_chars(i,word)
    l = ["N" for i in range(len(word))]
    for i in range(len(guess)):
        if guess[i] not in d:
            l[i] = "N"
        elif d[guess[i]] == 0:
            l[i] = "N"
        elif guess[i] == word[i]:
            l[i] = "H"
            d[guess[i]] -= 1
        elif guess[i] != word[i] and char_in_word(guess[i],word) == True:
            l[i] = "B"
            d[guess[i]] -= 1
    return l
    
def end_game_conditions(l,round_number, rounds):
    if round_number == rounds:
        return True
    for i in l:
        if i != "H":
            return False
    
    print("Congrats you got it!")
    return True
    
def not_all_letters(word):
    if type(word) == "NoneType" or word == None:
        return True
    word = word.lower()
    for i in word:
        if i < 'a' or i > 'z':
            return True
    
    return False
       
def has_duplicates(word):
    if type(word) == "NoneType" or word == None:
        return False
    d = {}
    word = word.lower()
    for i in word:
        if i in d:
            return True
        else:
            d[i] = 1
    
    return False
    

def main():
    r = RandomWords()
    rounds = 10
    round_num = 1
    disable_duplicates = True
    reshuffle = False
    
    duplicate_status = "Duplicates Disabled" if disable_duplicates == True else "Duplicates Enabled"
    
    MIN_LEN = 3
    MAX_LEN = 7
    
    custom_words = ["hello", "goodbye", "yes"]
    
    word = "kjlsdaf"
    while (word == "kjlsdaf") or (type(word) == "NoneType") or (disable_duplicates and (has_duplicates(word) == True)) or reshuffle or not_all_letters(word):
        word = r.get_random_word(excludePartOfSpeech="verb,preposition,conjunction,adverb",minLength=MIN_LEN, maxLength=MAX_LEN)
        if not ((word == None) or (type(word) == "NoneType")):
            z = zipf_frequency(word,'en')
            if z < 3:
                reshuffle = True
            else:
                reshuffle = False

    guess = ""
    hits_and_blows = []
    #word = custom_words[random.randint(0,len(custom_words))]
    #word = "gopala"
    rounds = len(word)*2
    print(str(len(word))+" letters, "+str(rounds)+" rounds, "+duplicate_status)
    word = word.lower()
    print(word)
    while round_num <= rounds:
        while (len(guess) != len(word)) or (is_all_lower(guess) == False):
            guess = str(input("Round "+str(round_num)+" of "+str(rounds)+": Please input guess: "))
            if (len(guess) != len(word)) or (is_all_lower(guess) == False):
                print("Please enter a valid guess (must have valid characters and be all lowercase): ")
        
        hits_and_blows = hit_blow_neither(guess,word)
        
        print([char for char in guess])
        print(hits_and_blows)
        
        if end_game_conditions(hits_and_blows, round_num, rounds) == True:
            print("The Game has ended")
            print("The word was: "+word)
            break
        round_num += 1
        guess = ""

main()
    

```
